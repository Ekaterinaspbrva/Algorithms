```python
import pandas as pd

data = {
    'name': ['Ğ˜Ğ²Ğ°Ğ½', 'ĞœĞ°Ñ€Ğ¸Ñ'],
    'age': [25, 30]
}

df = pd.DataFrame(data)
print(df)
```

```python
# ğŸ¼ Pandas: ĞŸÑ€ĞµĞ´Ğ¾Ğ±Ñ€Ğ°Ğ±Ğ¾Ñ‚ĞºĞ° Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ… Ğ´Ğ»Ñ Ğ¼Ğ°ÑˆĞ¸Ğ½Ğ½Ğ¾Ğ³Ğ¾ Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ¸Ñ
import pandas as pd
from sklearn.preprocessing import StandardScaler

# 1. Ğ¡Ğ¾Ğ·Ğ´Ğ°Ğ½Ğ¸Ğµ Ğ¿Ñ€Ğ¸Ğ¼ĞµÑ€Ğ° Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ…
data = {
    'age': [25, 30, 35, None, 40, 30],
    'income': [50000, 60000, 80000, 75000, None, 60000],
    'gender': ['male', 'female', 'female', 'male', 'female', 'male'],
    'city': ['Moscow', 'Moscow', 'SPB', 'SPB', 'SPB', 'Moscow'],
    'target': [0, 1, 0, 1, 0, 1]
}

df = pd.DataFrame(data)
print("Ğ˜ÑÑ…Ğ¾Ğ´Ğ½Ñ‹Ğµ Ğ´Ğ°Ğ½Ğ½Ñ‹Ğµ:")
print(df)

# 2. ĞŸÑ€Ğ¾Ğ²ĞµÑ€ĞºĞ° Ğ½Ğ° Ğ¿Ñ€Ğ¾Ğ¿ÑƒÑĞºĞ¸
print("\nĞŸÑ€Ğ¾Ğ¿ÑƒÑ‰ĞµĞ½Ğ½Ñ‹Ğµ Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ¸Ñ:")
print(df.isna().sum())

# 3. Ğ—Ğ°Ğ¿Ğ¾Ğ»Ğ½ĞµĞ½Ğ¸Ğµ Ğ¿Ñ€Ğ¾Ğ¿ÑƒÑ‰ĞµĞ½Ğ½Ñ‹Ñ… Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ¸Ğ¹
# Ğ’Ğ¾Ğ·Ñ€Ğ°ÑÑ‚ â€” ÑÑ€ĞµĞ´Ğ½ĞµĞµ, Ğ´Ğ¾Ñ…Ğ¾Ğ´ â€” Ğ¼ĞµĞ´Ğ¸Ğ°Ğ½Ğ°
df['age'] = df['age'].fillna(df['age'].mean())
df['income'] = df['income'].fillna(df['income'].median())

# 4. ĞšĞ¾Ğ´Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ¸Ğµ ĞºĞ°Ñ‚ĞµĞ³Ğ¾Ñ€Ğ¸Ğ°Ğ»ÑŒĞ½Ñ‹Ñ… Ğ¿Ñ€Ğ¸Ğ·Ğ½Ğ°ĞºĞ¾Ğ²
# Gender â€” Ğ·Ğ°Ğ¼ĞµĞ½Ğ° male/female Ğ½Ğ° 0/1
df['gender_encoded'] = df['gender'].map({'male': 0, 'female': 1})

# City â€” One-Hot Encoding
df = pd.get_dummies(df, columns=['city'])
print("\nĞšĞ¾Ğ´Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ½Ñ‹Ğµ Ğ´Ğ°Ğ½Ğ½Ñ‹Ğµ:")
print(df)


# 5. ĞœĞ°ÑÑˆÑ‚Ğ°Ğ±Ğ¸Ñ€Ğ¾Ğ²Ğ°Ğ½Ğ¸Ğµ Ñ‡Ğ¸ÑĞ»Ğ¾Ğ²Ñ‹Ñ… Ğ¿Ñ€Ğ¸Ğ·Ğ½Ğ°ĞºĞ¾Ğ²
scaler = StandardScaler()
df[['age', 'income']] = scaler.fit_transform(df[['age', 'income']])

# 6. Ğ Ğ°Ğ·Ğ´ĞµĞ»ĞµĞ½Ğ¸Ğµ X Ğ¸ y
X = df.drop(['target', 'gender'], axis=1)
y = df['target']

print("\nĞĞ±Ñ€Ğ°Ğ±Ğ¾Ñ‚Ğ°Ğ½Ğ½Ñ‹Ğµ Ğ¿Ñ€Ğ¸Ğ·Ğ½Ğ°ĞºĞ¸ (X):")
print(X.head())
print("\nĞ¦ĞµĞ»ĞµĞ²Ğ°Ñ Ğ¿ĞµÑ€ĞµĞ¼ĞµĞ½Ğ½Ğ°Ñ (y):")
print(y.head())
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# 1. Ğ¡Ğ¾Ğ·Ğ´Ğ°Ğ½Ğ¸Ğµ Ğ¿Ñ€Ğ¾ÑÑ‚Ğ¾Ğ¹ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X.shape[1],)),
    layers.Dense(1, activation='sigmoid')  # Ğ´Ğ»Ñ Ğ±Ğ¸Ğ½Ğ°Ñ€Ğ½Ğ¾Ğ¹ ĞºĞ»Ğ°ÑÑĞ¸Ñ„Ğ¸ĞºĞ°Ñ†Ğ¸Ğ¸
])

# 2. ĞšĞ¾Ğ¼Ğ¿Ğ¸Ğ»ÑÑ†Ğ¸Ñ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# 3. ĞĞ±ÑƒÑ‡ĞµĞ½Ğ¸Ğµ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸
model.fit(X, y, epochs=10, batch_size=4, validation_split=0.2)

# 4. ĞÑ†ĞµĞ½ĞºĞ°
loss, acc = model.evaluate(X, y)
print(f"Ğ¢Ğ¾Ñ‡Ğ½Ğ¾ÑÑ‚ÑŒ: {acc:.2f}")
```

```python
# 7. Ğ¡Ğ¾Ğ·Ğ´Ğ°Ğ½Ğ¸Ğµ Ğ¸ Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ¸Ğµ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸ Ñ TensorFlow
model = keras.Sequential([
    layers.Dense(16, activation='relu', input_shape=(X.shape[1],)),
    layers.Dense(8, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(X, y, epochs=20, batch_size=2, verbose=1)

# 8. ĞÑ†ĞµĞ½ĞºĞ° Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸
loss, acc = model.evaluate(X, y, verbose=0)
print(f"\nğŸ”¢ Ğ¢Ğ¾Ñ‡Ğ½Ğ¾ÑÑ‚ÑŒ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸: {acc:.2f}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# 9. ĞŸÑ€Ğ¸Ğ¼ĞµÑ€ Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ¸Ñ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸ Ğ½Ğ° PyTorch
# ĞŸÑ€Ğ¾ÑÑ‚ĞµĞ¹ÑˆĞ°Ñ Ñ€ĞµĞ³Ñ€ĞµÑÑĞ¸Ğ¾Ğ½Ğ½Ğ°Ñ Ğ·Ğ°Ğ´Ğ°Ñ‡Ğ°: y = 2x
X_pt = torch.tensor([[0.0], [1.0], [2.0], [3.0]])
y_pt = torch.tensor([[0.0], [2.0], [4.0], [6.0]])

class SimpleModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(1, 1)

    def forward(self, x):
        return self.linear(x)

model_pt = SimpleModel()
criterion = nn.MSELoss()
optimizer = optim.SGD(model_pt.parameters(), lr=0.01)

for epoch in range(100):
    y_pred = model_pt(X_pt)
    loss = criterion(y_pred, y_pt)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f"[PyTorch] Ğ­Ğ¿Ğ¾Ñ…Ğ° {epoch}: ĞŸĞ¾Ñ‚ĞµÑ€Ğ¸ = {loss.item():.4f}")
        print("[PyTorch] ĞŸÑ€ĞµĞ´ÑĞºĞ°Ğ·Ğ°Ğ½Ğ¸Ğµ Ğ¿Ñ€Ğ¸ x=4.0:", model_pt(torch.tensor([[4.0]])).item())

```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
# 10. ĞŸÑ€Ğ¸Ğ¼ĞµÑ€ Ñ Ğ¸ÑĞ¿Ğ¾Ğ»ÑŒĞ·Ğ¾Ğ²Ğ°Ğ½Ğ¸ĞµĞ¼ Scikit-learn
clf = LogisticRegression()
clf.fit(X, y)
y_pred = clf.predict(X)
acc_sklearn = accuracy_score(y, y_pred)
print(f"\nğŸ¯ Ğ¢Ğ¾Ñ‡Ğ½Ğ¾ÑÑ‚ÑŒ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸ Scikit-learn (LogisticRegression): {acc_sklearn:.2f}")
```

```python
# 11. ĞŸÑ€Ğ¸Ğ¼ĞµÑ€Ñ‹ Ñ XGBoost / LightGBM / CatBoost

from xgboost import XGBClassifier, plot_importance
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import seaborn as sns

xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X, y)
y_xgb = xgb_model.predict(X)
print(f"ğŸ¹ XGBoost Accuracy: {accuracy_score(y, y_xgb):.2f}")

lgb_model = LGBMClassifier()
lgb_model.fit(X, y)
y_lgb = lgb_model.predict(X)
print(f"ğŸ’¡ LightGBM Accuracy: {accuracy_score(y, y_lgb):.2f}")

cat_model = CatBoostClassifier(verbose=0)
cat_model.fit(X, y)
y_cat = cat_model.predict(X)
print(f"ğŸ± CatBoost Accuracy: {accuracy_score(y, y_cat):.2f}")
```

```python
# 12. ĞšÑ€Ğ¾ÑÑ-Ğ²Ğ°Ğ»Ğ¸Ğ´Ğ°Ñ†Ğ¸Ñ Ğ¼Ğ¾Ğ´ĞµĞ»ĞµĞ¹
cv_scores = {
    'LogisticRegression': cross_val_score(clf, X, y, cv=3).mean(),
    'XGBoost': cross_val_score(xgb_model, X, y, cv=3).mean(),
    'LightGBM': cross_val_score(lgb_model, X, y, cv=3).mean(),
    'CatBoost': cross_val_score(cat_model, X, y, cv=3).mean()
}

print("\nğŸ“Š Ğ¡Ñ€ĞµĞ´Ğ½ÑÑ Ñ‚Ğ¾Ñ‡Ğ½Ğ¾ÑÑ‚ÑŒ Ğ½Ğ° ĞºÑ€Ğ¾ÑÑ-Ğ²Ğ°Ğ»Ğ¸Ğ´Ğ°Ñ†Ğ¸Ğ¸:")
for model_name, score in cv_scores.items():
    print(f"{model_name}: {score:.2f}")

# 13. Ğ’Ğ¸Ğ·ÑƒĞ°Ğ»Ğ¸Ğ·Ğ°Ñ†Ğ¸Ñ Ğ²Ğ°Ğ¶Ğ½Ğ¾ÑÑ‚Ğ¸ Ğ¿Ñ€Ğ¸Ğ·Ğ½Ğ°ĞºĞ¾Ğ² (XGBoost)
plt.figure(figsize=(8, 5))
plot_importance(xgb_model)
plt.title("Ğ’Ğ°Ğ¶Ğ½Ğ¾ÑÑ‚ÑŒ Ğ¿Ñ€Ğ¸Ğ·Ğ½Ğ°ĞºĞ¾Ğ² (XGBoost)")
plt.show()

# 14. Ğ“Ñ€Ğ°Ñ„Ğ¸Ğº ÑÑ€Ğ°Ğ²Ğ½ĞµĞ½Ğ¸Ñ Ñ‚Ğ¾Ñ‡Ğ½Ğ¾ÑÑ‚Ğ¸ Ğ¼Ğ¾Ğ´ĞµĞ»ĞµĞ¹
plt.figure(figsize=(8, 5))
sns.barplot(x=list(cv_scores.keys()), y=list(cv_scores.values()))
plt.title("Ğ¡Ñ€Ğ°Ğ²Ğ½ĞµĞ½Ğ¸Ğµ Ğ¼Ğ¾Ğ´ĞµĞ»ĞµĞ¹ Ğ¿Ğ¾ ÑÑ€ĞµĞ´Ğ½ĞµĞ¹ Ñ‚Ğ¾Ñ‡Ğ½Ğ¾ÑÑ‚Ğ¸ (CV)")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.show()
```
