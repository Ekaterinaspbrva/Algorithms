```python
# 🐼 Pandas: Предобработка данных для машинного обучения
import pandas as pd
from sklearn.preprocessing import StandardScaler

# 1. Создание примера данных
data = {
    'age': [25, 30, 35, None, 40, 30],
    'income': [50000, 60000, 80000, 75000, None, 60000],
    'gender': ['male', 'female', 'female', 'male', 'female', 'male'],
    'city': ['Moscow', 'Moscow', 'SPB', 'SPB', 'SPB', 'Moscow'],
    'target': [0, 1, 0, 1, 0, 1]
}

df = pd.DataFrame(data)
print("Исходные данные:")
print(df)

# 2. Проверка на пропуски
print("\nПропущенные значения:")
print(df.isna().sum())

# 3. Заполнение пропущенных значений
# Возраст — среднее, доход — медиана
df['age'] = df['age'].fillna(df['age'].mean())
df['income'] = df['income'].fillna(df['income'].median())

# 4. Кодирование категориальных признаков
# Gender — замена male/female на 0/1
df['gender_encoded'] = df['gender'].map({'male': 0, 'female': 1})

# City — One-Hot Encoding
df = pd.get_dummies(df, columns=['city'])
print("\nКодированные данные:")
print(df)


# 5. Масштабирование числовых признаков
scaler = StandardScaler()
df[['age', 'income']] = scaler.fit_transform(df[['age', 'income']])

# 6. Разделение X и y
X = df.drop(['target', 'gender'], axis=1)
y = df['target']

print("\nОбработанные признаки (X):")
print(X.head())
print("\nЦелевая переменная (y):")
print(y.head())
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# 1. Создание простой модели
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X.shape[1],)),
    layers.Dense(1, activation='sigmoid')  # для бинарной классификации
])

# 2. Компиляция модели
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# 3. Обучение модели
model.fit(X, y, epochs=10, batch_size=4, validation_split=0.2)

# 4. Оценка
loss, acc = model.evaluate(X, y)
print(f"Точность: {acc:.2f}")
```

```python
# 7. Создание и обучение модели с TensorFlow
model = keras.Sequential([
    layers.Dense(16, activation='relu', input_shape=(X.shape[1],)),
    layers.Dense(8, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(X, y, epochs=20, batch_size=2, verbose=1)

# 8. Оценка модели
loss, acc = model.evaluate(X, y, verbose=0)
print(f"\n🔢 Точность модели: {acc:.2f}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# 9. Пример обучения модели на PyTorch
# Простейшая регрессионная задача: y = 2x
X_pt = torch.tensor([[0.0], [1.0], [2.0], [3.0]])
y_pt = torch.tensor([[0.0], [2.0], [4.0], [6.0]])

class SimpleModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(1, 1)

    def forward(self, x):
        return self.linear(x)

model_pt = SimpleModel()
criterion = nn.MSELoss()
optimizer = optim.SGD(model_pt.parameters(), lr=0.01)

for epoch in range(100):
    y_pred = model_pt(X_pt)
    loss = criterion(y_pred, y_pt)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f"[PyTorch] Эпоха {epoch}: Потери = {loss.item():.4f}")
        print("[PyTorch] Предсказание при x=4.0:", model_pt(torch.tensor([[4.0]])).item())

```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
# 10. Пример с использованием Scikit-learn
clf = LogisticRegression()
clf.fit(X, y)
y_pred = clf.predict(X)
acc_sklearn = accuracy_score(y, y_pred)
print(f"\n🎯 Точность модели Scikit-learn (LogisticRegression): {acc_sklearn:.2f}")
```

```python
# 11. Примеры с XGBoost / LightGBM / CatBoost

from xgboost import XGBClassifier, plot_importance
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import seaborn as sns

xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X, y)
y_xgb = xgb_model.predict(X)
print(f"🏹 XGBoost Accuracy: {accuracy_score(y, y_xgb):.2f}")

lgb_model = LGBMClassifier()
lgb_model.fit(X, y)
y_lgb = lgb_model.predict(X)
print(f"💡 LightGBM Accuracy: {accuracy_score(y, y_lgb):.2f}")

cat_model = CatBoostClassifier(verbose=0)
cat_model.fit(X, y)
y_cat = cat_model.predict(X)
print(f"🐱 CatBoost Accuracy: {accuracy_score(y, y_cat):.2f}")
```

```python
# 12. Кросс-валидация моделей
cv_scores = {
    'LogisticRegression': cross_val_score(clf, X, y, cv=3).mean(),
    'XGBoost': cross_val_score(xgb_model, X, y, cv=3).mean(),
    'LightGBM': cross_val_score(lgb_model, X, y, cv=3).mean(),
    'CatBoost': cross_val_score(cat_model, X, y, cv=3).mean()
}

print("\n📊 Средняя точность на кросс-валидации:")
for model_name, score in cv_scores.items():
    print(f"{model_name}: {score:.2f}")

# 13. Визуализация важности признаков (XGBoost)
plt.figure(figsize=(8, 5))
plot_importance(xgb_model)
plt.title("Важность признаков (XGBoost)")
plt.show()

# 14. График сравнения точности моделей
plt.figure(figsize=(8, 5))
sns.barplot(x=list(cv_scores.keys()), y=list(cv_scores.values()))
plt.title("Сравнение моделей по средней точности (CV)")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.show()
```

```python
import pandas as pd
from datetime import datetime


# Создание данных
data = {
    "customer_id": [1, 2, 3, 4, 5],
    "name": ["Иван Иванов", "Мария Петрова", "Игорь Смирнов", "Анна Кузнецова", "Иван Иванов"],
    "age": [25, None, 35, 28, 25],
    "email": ["ivan@example.com", "maria@", "igor@example.com", "anna@example.com", "ivan@example.com"],
    "signup_date": ["2023-01-15", "2023-03-22", "2023-02-10", "not_a_date", "2023-01-15"],
    "purchase_amount": [100.0, 200.5, None, 150.0, 100.0]
}


# Преобразование в DataFrame
df = pd.DataFrame(data)

# Сохранение в файл
df.to_csv("customers.csv", index=False)

# Загрузка для проверки
df_loaded = pd.read_csv("customers.csv")
print("✅ Файл создан и загружен:")
df_loaded



```

```python
# 1. Удаление дубликатов
df = df.drop_duplicates()

# 2. Заполнение пропущенных значений в 'age' средним значением
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['age'] = df['age'].fillna(df['age'].mean())

# 3. Обработка пропущенных значений в 'purchase_amount'
df['purchase_amount'] = df['purchase_amount'].fillna(0)

# 4. Очистка email — убираем записи, где email не содержит '@'
df = df[df['email'].str.contains('@', na=False)]

# 5. Преобразование signup_date в формат datetime
df['signup_date'] = pd.to_datetime(df['signup_date'], errors='coerce')

# 6. Удаление строк с некорректными датами
df = df.dropna(subset=['signup_date'])

# Финальный результат
print(df)
```
